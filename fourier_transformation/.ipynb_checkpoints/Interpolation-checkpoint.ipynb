```python
import pandas as pd

url ='https://archive.stsci.edu/hlsps/k2sff/c03/206100000/03150/hlsp_k2sff_k2_lightcurve_206103150-c03_kepler_v1_llc-default-aper.txt'
df = pd.read_csv(url, skiprows=0, header=0, sep=',', index_col=False)
display(df)
```

```python
arr = df[df.columns[1]]
display(arr)

xp = df[df.columns[0]] # X coordinates of arr
#plt.scatter(xp, arr)
#plt.show()

spacing = 0.002
targets = np.arange(xp[0], xp[len(arr) -1]-spacing, spacing) # X coordinates desired
interpolated = np.interp(targets, xp, arr) # Interpolate new data points
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
#ax.scatter(targets, interpolated)
#plt.show()

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlim((2149.75, 2150.25))
# fig = ax.set_ylim((np.min(y_points) - 2, np.max(y_points) + 2))
#ax.scatter(targets, interpolated)
#plt.show()
```

```python
print("signal")
# Signalet
x = df[df.columns[0]]
y = df[df.columns[1]]

# Signalet vrides rundt
r_cord = []
min_freq_range = 0.0
max_freq_range = 10.0

sf_list = np.arange(min_freq_range, max_freq_range, 0.01)
for sf in sf_list:
    r_cord.append([(y[i], x[i]*sf*2*np.pi) for i in range(len(x))])

x_cord, y_cord = [], []
for l in range(len(r_cord)):
    x_cord.append( [amp * np.cos(theta) for (amp, theta) in r_cord[l]] )
    y_cord.append( [amp * np.sin(theta) for (amp, theta) in r_cord[l]] ) 
print("done")
```

```python
print("mean")
# Calculate mean values
x_mean_list = []
y_mean_list = []
amp_list = []
phase_list = []

plt.rcParams["figure.figsize"] = (15,110)
for l in range(len(r_cord)):
    """plt.subplot(int(len(r_cord)/4)+1, 4, int(l+1))
    plt.plot(x_cord[l], y_cord[l])
    plt.plot(np.mean(x_cord[l]), np.mean(y_cord[l]), 'or' )
    plt.title("Sampling Freq = "+str(round(sf_list[l], 2))+" Hz")
    """
    # Storing the COM for plotting later
    x_mean = np.sum(x_cord[l])
    y_mean = np.sum(y_cord[l])
    x_mean_list.append(x_mean)
    y_mean_list.append(y_mean)
    amp_list.append(np.sqrt(x_mean**2 + y_mean**2))
    if x_mean == 0:
        phase_list.append(0)
    else:
        phase_list.append(np.arctan(y_mean/x_mean))
print("mean done")
```

```python
print("plot")
# Plotting
plt.rcParams["figure.figsize"] = (12,10)
#plt.plot(sf_list, x_mean_list)
#plt.plot(sf_list, y_mean_list, "C2")

#plt.plot(sf_list, np.add(x_mean_list, y_mean_list))
#plt.plot(sf_list, amp_list)
#plt.plot(sf_list, phase_list)

max_val = np.argmax(amp_list)
print(max_val)
print("{}: {}, {}".format(sf_list[max_val], amp_list[max_val], phase_list[max_val]))
plt.show()
print("plot done")
```

```python
# Smooth
min_freq = 0.5
max_freq = 5
print("smooth")
x_smoothed = [i if i>0 and i>0.01*max(x_mean_list) else 0 for i in x_mean_list]
y_smoothed = [i if i>0 and i>0.2*max(y_mean_list) else 0 for i in y_mean_list]
pyth = [np.sqrt(x_mean_list[i]**2 + y_mean_list[i]**2) for i in range(len(x_mean_list))]

min_index = 0
max_index = len(sf_list) - 1

for i in range(len(sf_list)):    
    if sf_list[i] >= max_freq and max_index == len(sf_list) - 1:
        max_index = i
        break
    elif sf_list[i] >= min_freq and min_index == 0:
        min_index = i     

sf_list_cut = sf_list[min_index:max_index]
x_smoothed_cut = x_smoothed[min_index:max_index]
y_smoothed_cut = y_smoothed[min_index:max_index]
    
# print(sf_list)
# x_smoothed = x_smoothed[1:]
"""print(max(x_smoothed))
print(max(y_smoothed))
print(max(y_smoothed) - max(x_smoothed))
shift = np.arctan2(max(y_smoothed), max(x_smoothed))
print(shift/2*np.pi)
"""
plt.rcParams["figure.figsize"] = (20, 10)
fig3=plt.figure()
ax3=fig3.add_axes([0,0,1,1])
ax3.plot(sf_list_cut, x_smoothed_cut)

#plt.rcParams["figure.figsize"] = (12,10)
ax3.plot(sf_list_cut, x_smoothed_cut)
ax3.plot(sf_list_cut, y_smoothed_cut)
plt.show()
# plt.plot(sf_list, pyth)
#plt.xlabel("Frequency [Hz]")
#plt.xticks(np.arange(min(sf_list_cut), max(sf_list_cut), 0.5))
# plt.show()

# Good frequencies
freq_idx = np.argmax(x_smoothed_cut)
print([sf_list_cut[freq_idx], complex(x_smoothed_cut[freq_idx], y_smoothed_cut[freq_idx])])
freq_idx = np.argmax(y_smoothed_cut)
print([sf_list_cut[freq_idx], complex(x_smoothed_cut[freq_idx], y_smoothed_cut[freq_idx])])

print("Delta tid: {}".format(datetime.now() - initial_time))
```

```python
#interactive plotting in separate window
%matplotlib inline 
# %matplotlib tk

min_fold_freq = 4
max_fold_freq = 4.2
fold_freqs = np.arange(min_fold_freq, max_fold_freq, 0.05)

plt.rcParams["figure.figsize"] = (20,40)
for freq_idx in range(len(fold_freqs)):
    plt.subplot(int(len(fold_freqs)/2)+1, 4, int(freq_idx+1))    
    folding_freq = fold_freqs[freq_idx]
    
    data_folding = [[(2 + xp[i]) % folding_freq, arr[i]] for i in range(len(xp))]
    data_folding_x = []
    data_folding_y = []
    for point in data_folding:
        data_folding_x.append(point[0])
        data_folding_y.append(point[1])
    #print(data_folding_x)
    plt.scatter(data_folding_x, data_folding_y)
    plt.title("Folding Freq = "+str(round(folding_freq, 5))+" Hz")
plt.show()

```
